# FB ads

```python
import pandas as pd

df = pd.read_csv('/Users/pranavdalvi/Research Analyst ischool/Task_03_Descriptive_Stats/facebook_ads_cleaned.csv')

# General stats
print(df.describe(include='all'))

# Value counts for non-numeric
for col in df.select_dtypes(include='object'):
    print(col)
    print(df[col].value_counts().head(5))

# Group by page_id
print(df.groupby('page_id').describe())

# Group by page_id and ad_id
print(df.groupby(['page_id', 'ad_id']).describe())

```

# FB posts

```python
import pandas as pd

# Load dataset
df = pd.read_csv("/Users/pranavdalvi/Research Analyst ischool/Task_03_Descriptive_Stats/facebook_posts_cleaned.csv")

# List of numeric columns to clean and analyze
numeric_cols = [
    "Total Interactions", "Likes", "Comments", "Shares", "Love", "Wow", "Haha",
    "Sad", "Angry", "Care", "Post Views", "Total Views",
    "Total Views For All Crossposts", "Overperforming Score"
]

# Clean numeric columns (remove commas and convert to numbers)
for col in numeric_cols:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col].astype(str).str.replace(",", ""), errors='coerce')

# Print descriptive statistics
print("\n=== Descriptive Statistics ===")
print(df[numeric_cols].describe())

# Value counts for common categorical columns
categorical_cols = ["Type", "Video Share Status", "Is Video Owner?"]
print("\n=== Categorical Column Value Counts ===")
for col in categorical_cols:
    if col in df.columns:
        print(f"\n{col}:\n{df[col].value_counts().head()}")
```

# Twitter Posts

```python
import pandas as pd

# Load dataset
df = pd.read_csv("/Users/pranavdalvi/Research Analyst ischool/Task_03_Descriptive_Stats/twitter_posts_cleaned.csv")

# Clean numeric columns
numeric_cols = [
    "retweetCount", "replyCount", "likeCount", "quoteCount", "viewCount",
    "bookmarkCount"
]

for col in numeric_cols:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col].astype(str).str.replace(",", ""), errors="coerce")

# Compute descriptive statistics
print("\n=== Descriptive Statistics ===")
print(df[numeric_cols].describe())

# Categorical value counts
categorical_cols = ["source", "lang", "isReply", "isQuote", "isRetweet"]
print("\n=== Categorical Column Value Counts ===")
for col in categorical_cols:
    if col in df.columns:
        print(f"\n{col}:\n{df[col].value_counts().head()}")
```
